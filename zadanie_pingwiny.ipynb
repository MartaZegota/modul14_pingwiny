Import bazy danych

```python
import pandas as pd
ping = pd.read_csv('penguins.csv')
ping
```

Ustalenie zmiennej objaśnianej y, którą będzie Species. Pozostałe to zmienne objaśniające X.

```python
features = ['CulmenLength','CulmenDepth','FlipperLength','BodyMass']
target = 'Species'
X, y = ping[features], ping[target]
print('X:')
display(X)
print('\ny:')
display(y)
```

Ustalenie ilości klas zmiennej y i procentowego udziału ich wystąpień w bazie

```python
pd.Series(y).value_counts(normalize=True)
```

Sprawdzenie typów danych i wartości brakujących

```python
ping.info()
```

Dane są wartościami numerycznymi

Sprawdzenie rozkładów zmiennych X

```python
X.describe()
```

Porównanie zmiennych numerycznych X do zmiennej kategorycznej y z wykorzystaniem wykresów pudełkowych

```python
import matplotlib.pyplot as plt
for col in features:
    print(f'Unique values: {len(ping[col].unique())}')
    ping.boxplot(column=col, by='Species', figsize=(6,5))
    plt.title(col)
    plt.show()
```

Z powyższych wizualizacji wynika, że można szacunkowo wskazać gatunek 0 pingwina po jego długości dzioba ('CulmenLength'), który jest krótszy od pozostałych dwóch gatunków. Natomiast gatunek 1 pingwina można estymować po głębokości dzioba ('CulmenDepth'), który jest płytszy od pozostałych dwóch gatunków oraz po długości płetw ('FlipperLength'), które są wyraźnie większe i masie ciała ('BodyMass'), która jest również większa od pozostałych dwóch gatunków.

Porównanie zmiennych numerycznych X do zmiennej kategorycznej y z wykorzystaniem histogramów

```python
import seaborn as sns
for col in features:
    plt.figure(figsize=(12, 2.5))
    sns.distplot(ping.loc[ping['Species']==0, col], kde=True, rug=False, color='green', label='Gatunek 0')
    sns.distplot(ping.loc[ping['Species']==1, col], kde=True, rug=False, color='red', label='Gatunek 1')
    sns.distplot(ping.loc[ping['Species']==2, col], kde=True, rug=False, color='yellow', label='Gatunek 2')
    plt.legend(loc='upper right')
    plt.show()
```

Na histogramach widać jak zmienia się gęstość klas zmiennej objaśnianej y w zależności od wartości poszczególnych cech X.
Wnioski z analizy histogramów są zbieżne z tymi wyciągniętymi dla wykresów pudełkowych.

Korelacja zmiennych numerycznych (objaśniających) X

```python
plt.figure(figsize=(8, 8))
ax = sns.heatmap(X.corr(),
                 xticklabels=X.corr().columns,
                 yticklabels=X.corr().columns,
                 cmap='RdYlGn',
                 center=0,
                 annot=True)

plt.title('Korelacja dla zmiennych numerycznych X', fontsize=22)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Z powyższej grafiki wynika, że najmocniej korelują ze sobą długość płetw ('FlipperLength') i masa ciała ('BodyMass').

Z histogramów i wykresów pudełkowych można wywnioskować, że każda z 4 cech definuje w pewnym stopniu gatunek pingwina.

Accuracy: W powyższym badaniu dla większości klasyfikatorów dokładność jest niska. Jednocześnie zbiór stał się niezbalansowany, co oznacza, że zaczęła dominować jedna z trzech klas, co widać to na wizualizacjach. Wyjąrtkiem jest regresja logistyczna.

Precision: Wartość metryki precyzji jest bardzo niska z wyjątkiem regresji logistycznej.

Recall: Wynosi 1 w metrykach modeli niezbalansowanych.

F1 score: Metryka F1, będąca średnią harmoniczną precyzyji i czułości, jest najwyższa w regresji logistycznej.
